```python
student = pd.DataFrame({'key2': ['K0', 'K0', 'K1', 'K2'],
                     'student_id': ['ikd', 'uyo', 'lagos', 'benin'],
                        'student': ['A0', 'A1', 'A2', 'A3'],
                        'teacher': ['B0', 'B1', 'B2', 'B3']},[0,1,2,4])

student_grade = pd.DataFrame({'student_id': ['ikd', 'uyo', 'ikeja', 'benin'],
                               'action': ['K0', 'K0', 'K0', 'K0'],
                                  'grade': ['C0', 'C1', 'C2', 'C3'],
                                  'class': ['D0', 'D1', 'D2', 'D3']},[0,1,2,3])
```

```python
#merge
pd.merge(student, student_grade, how='left', on='student_id')#includes everything on left table and excludes what does not march on the right table
```

```python
pd.merge(student, student_grade, how='right', on='student_id')#includes everything on right table and excludes what does not march on the left table
```

```python
pd.merge(student, student_grade, how='inner', on='student_id')#includes only what is on both tables
```

```python
pd.merge(student, student_grade, how='outer', on='student_id')#everything both tables and what does not march NaN
```

```python
student.join(student_grade,how="right",lsuffix='student_id',rsuffix='student_id')
```

```python
pd.concat([student,student_grade],axis=0)
```
